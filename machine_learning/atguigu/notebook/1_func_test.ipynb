函数测试

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 定义函数y=sinx
x = np.arange(0, 6, 0.1)
y = np.sin(x)
print(np.array([[1,2,3,4,5.5,6],[11,12,13,41,15.5,16],[12,22,23,42,25.5,26]]))
```

```python
print(x)
```

```python
print(y)
```

```python
# 画出函数图像
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = sin(x)")
plt.show()
```

```python
# 导函数y' = cosx
y1 = np.cos(x)
print(y1)
```

```python
# 画出导函数图像
plt.plot(x, y1)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = cos(x)")
plt.show()
```

```python
# 将函数和导函数图像放在一张图中
plt.plot(x, y, label="y=sin(x)")
plt.plot(x, y1, label="y'=cos(x)", linestyle="--")
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = sin(x)")
plt.legend()
plt.show()
```
