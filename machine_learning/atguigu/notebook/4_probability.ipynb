```python
import numpy as np

# Generate a single random number from a standard normal distribution (mean=0, std=1)
random_number = np.random.normal()
print(f"Single random number: {random_number}")

# Generate an array of 10 random numbers from a normal distribution with mean 5 and std 2
samples = np.random.normal(loc=5, scale=2, size=10)
print(f"Array of samples: {samples}")

# Generate a 2x3 array of samples
matrix_samples = np.random.normal(loc=10, scale=3, size=(2, 3))
print(f"Matrix of samples:\n{matrix_samples}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Define distribution parameters
mean = 0
std_dev = 1

# 2. Generate x-values
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 500)

# 3. Calculate y-values (Probability Density Function)
y = norm.pdf(x, loc=mean, scale=std_dev)

# 4. Plot the distribution
plt.plot(x, y, label=f'μ = {mean}, σ = {std_dev}')

# 5. Customize the plot (optional)
plt.title('Normal Distribution')
plt.xlabel('X-axis')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# 6. Display the plot
plt.show()
```
