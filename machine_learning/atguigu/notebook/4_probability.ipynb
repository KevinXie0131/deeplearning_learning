```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Define distribution parameters
mean = 0
std_dev = 1

# 2. Generate x-values
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 500)

# 3. Calculate y-values (Probability Density Function)
y = norm.pdf(x, loc=mean, scale=std_dev)

# 4. Plot the distribution
plt.plot(x, y, label=f'μ = {mean}, σ = {std_dev}')

# 5. Customize the plot (optional)
plt.title('Normal Distribution')
plt.xlabel('X-axis')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# 6. Display the plot
plt.show()
```
